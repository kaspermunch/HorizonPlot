```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import mpld3
```

```python
%matplotlib inline
```

```python
n = 1000
pops = 50

df = pd.DataFrame({'chrom': ['chr1']*pops*n,
                'pop': [x for y in ([chr(65+i)]*n for i in range(pops)) for x in y],
                'start': list(range(1*n)) * pops, 
#                    'pi': list(np.sin(np.linspace(-np.pi, 10*np.pi, 1*n))+0.1) * pops })
                 'pi': np.add(list(np.sin(np.linspace(-np.pi, 10*np.pi, 1*n))) * pops, 5*np.random.random(n*pops))-2 })
```

```python
g = sns.FacetGrid(df, 
                    col='chrom', 
                    row='pop',size=0.3, aspect=50,
                    sharex=True)
g.map(plt.plot, 'start', 'pi') ;
```

```python
fig = horizon_plot(df, 'pi', width=1, col='chrom', row='pop', size=0.3, aspect=50)
#plt.savefig('tmp.pdf')
```
